## FIR & IIR BAND PASS

## FIR HAMMING WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter bandpass FIR
def design_bandpass_fir(numtaps, lowcut, highcut, fs):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    taps = signal.firwin(numtaps, [low, high], window='hamming', pass_zero=False)
    return taps

# Baca file WAV
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_FIR_HammingWindow_BandPass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
numtaps = 101  # Panjang filter
lowcut_freq = 500  # Frekuensi cutoff bawah dalam Hz
highcut_freq = 1500  # Frekuensi cutoff atas dalam Hz
taps = design_bandpass_fir(numtaps, lowcut_freq, highcut_freq, fs)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter Band-Pass')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

## FIR KAIZER WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter bandpass FIR dengan Kaiser window
def design_bandpass_fir_kaiser(numtaps, lowcut, highcut, fs, beta):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    taps = signal.firwin(numtaps, [low, high], window=('kaiser', beta), pass_zero=False)
    return taps

# Baca file WAV
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_FIR_KaiserWindow_BandPass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
numtaps = 101  # Panjang filter
lowcut_freq = 500  # Frekuensi cutoff bawah dalam Hz
highcut_freq = 1500  # Frekuensi cutoff atas dalam Hz
beta = 8.6  # Parameter beta untuk jendela Kaiser
taps = design_bandpass_fir_kaiser(numtaps, lowcut_freq, highcut_freq, fs, beta)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter Band-Pass (Kaiser Window)')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

## IIR CHEBYSHEV WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter bandpass IIR Chebyshev Type I
def design_bandpass_iir(lowcut, highcut, fs, order=4, rp=1):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.cheby1(order, rp, [low, high], btype='band')
    return b, a

# Baca file WAV
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_IIR_Chebyshev_BandPass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
lowcut_freq = 500  # Frekuensi cutoff bawah dalam Hz
highcut_freq = 1500  # Frekuensi cutoff atas dalam Hz
order = 4  # Orde filter
rp = 1  # Ripple dalam dB
b, a = design_bandpass_iir(lowcut_freq, highcut_freq, fs, order, rp)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter Band-Pass IIR Chebyshev')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
impulse = np.zeros(1024)
impulse[0] = 1
impulse_response = signal.lfilter(b, a, impulse)
plt.subplot(4, 1, 3)
plt.plot(impulse_response)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

## IIR ELLIPTIC WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter bandpass IIR Elliptic
def design_bandpass_iir(lowcut, highcut, fs, order=5, rp=1, rs=40):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.ellip(order, rp, rs, [low, high], btype='band', analog=False)
    return b, a

# Baca file WAV
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_IIR_Elliptic_BandPass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
lowcut_freq = 500  # Frekuensi cutoff bawah dalam Hz
highcut_freq = 1500  # Frekuensi cutoff atas dalam Hz
order = 5  # Orde filter
rp = 1  # Ripple di passband (dB)
rs = 40  # Penolakan di stopband (dB)
b, a = design_bandpass_iir(lowcut_freq, highcut_freq, fs, order, rp, rs)

# Terapkan filter ke sinyal
filtered_data = signal.filtfilt(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter Band-Pass IIR Elliptic')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
impulse = np.zeros(100)
impulse[0] = 1
impulse_response = signal.lfilter(b, a, impulse)
plt.subplot(4, 1, 3)
plt.plot(impulse_response)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```
