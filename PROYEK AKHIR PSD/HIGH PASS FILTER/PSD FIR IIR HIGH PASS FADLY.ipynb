## FIR & IIR HIGH PASS

## FIR Hamming

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter highpass FIR
def design_highpass_fir(numtaps, cutoff, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    taps = signal.firwin(numtaps, norm_cutoff, window='hamming', pass_zero=False)
    return taps

# Baca file WAV
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_FIR_HammingWindow.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
numtaps = 101  # Panjang filter
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
taps = design_highpass_fir(numtaps, cutoff_freq, fs)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

## IIR Chebyshev Window

## IIR Butterworth

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter highpass Butterworth IIR
def design_highpass_butter(order, cutoff, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    b, a = signal.butter(order, norm_cutoff, btype='high', analog=False)
    return b, a

# Baca file WAV
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_IIR_Butterworth.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
order = 5  # Orde filter
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
b, a = design_highpass_butter(order, cutoff_freq, fs)

# Terapkan filter ke sinyal
filtered_data = signal.filtfilt(b, a, data)

# Normalisasi sinyal terfilter
filtered_data = np.int16(filtered_data / np.max(np.abs(filtered_data)) * 32767)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data)

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
impulse = np.zeros_like(data)
impulse[0] = 1
response = signal.lfilter(b, a, impulse)
plt.plot(response)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

## IIR Elliptic

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Function to design a highpass elliptic IIR filter
def design_highpass_elliptic(order, cutoff, ripple_db, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    rp = ripple_db  # Passband ripple in dB
    rs = 60  # Stopband attenuation in dB
    b, a = signal.ellip(order, rp, rs, norm_cutoff, btype='high', analog=False)
    return b, a

# Read WAV file
input_wav_path = 'Wonderwall (Remastered) (320).wav'
output_wav_path = 'Output_IIR_Elliptic.wav'
fs, data = wavfile.read(input_wav_path)

# Ensure data is 1D array
if data.ndim > 1:
    data = data[:, 0]

# Design filter
order = 5  # Filter order
cutoff_freq = 1000  # Cutoff frequency in Hz
ripple_db = 1  # Passband ripple in dB
b, a = design_highpass_elliptic(order, cutoff_freq, ripple_db, fs)

# Apply filter to signal
filtered_data = signal.lfilter(b, a, data)

# Normalize filtered signal
filtered_data = np.int16(filtered_data / np.max(np.abs(filtered_data)) * 32767)

# Save filtered signal to WAV file
wavfile.write(output_wav_path, fs, filtered_data)

# Plot original and filtered signals
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.plot(data)
plt.title('Original Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(filtered_data)
plt.title('Filtered Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(3, 1, 3)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```
